# 19. Classify movies based on rating

```python
def mov_rating(rating):
  if rating>=7.5:
    return "Excellent"
  elif rating>=6.5 and rating<7.5:
    return "Good"
  elif rating>=5.5 and rating<6.5:
    return "Average"
  else:
    return "Poor"
data['Rating category']=data['Rating'].apply(mov_rating)
data
```
